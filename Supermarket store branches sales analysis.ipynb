# Step 2: Data Visualization:
 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Items_Available',y='Store_Area', data=df)
```

The above graph gives an indication about the constant proportionality of the store Area and the item availability. 

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Store_Sales',y='Store_Area',data=df,color="purple")
```

There is a weak positive, nonlinear association between the two variables, with an outlier.

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Daily_Customer_Count',y='Items_Available', data=df)
```

There is a weak negative, nonlinear association between the two variables, with an outlier.

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Store_Sales',y='Daily_Customer_Count', data=df)
```

There is no correlation association between the two variables and it's normally distibuted and concentrated at the centre , with an outlier.
